```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta


crypto_list = ["BNBUSDT"]
interval = '15m'
file_path = f'/Volumes/Ext-Disk/data/futures/um/daily/klines'

raw_data_all = {}
feed_data_all = {}
pnl_all = {}
fee_rate = 0.0001
suffix = "2025-01-01_2025-07-01"

def generate_date_range(start_date, end_date):    
    start = datetime.strptime(start_date, '%Y-%m-%d')
    end = datetime.strptime(end_date, '%Y-%m-%d')
    
    date_list = []
    current = start
    while current <= end:
        date_list.append(current.strftime('%Y-%m-%d'))
        current += timedelta(days=1)
    return date_list
```

```python
start_date = "2025-01-01"
end_date = "2025-06-30"

date_list = generate_date_range(start_date, end_date)
crypto_data = {}
crypto_date_data = []
for crypto in crypto_list:
    crypto_data[crypto] = {}
    crypto_date_data = []
    for date in date_list:
        pd_csv = pd.read_csv(f"{file_path}/{crypto}/{interval}/{suffix}/{crypto}-{interval}-{date}.zip")
        crypto_date_data.append(pd_csv)
    
    # crypto_data[crypto] = crypto_date_data

# print(crypto_data["BNBUSDT"])
```
